```python
from dataset import STCNetDataset
from torchvision import transforms
stc_dataset = STCNetDataset('data/dataset_v1/train/', None, transforms.Compose(
        [transforms.ToTensor()]))
```

```python
from torch.utils.data import DataLoader
train_loader = DataLoader(dataset=stc_dataset, batch_size=1, pin_memory=True)
```

```python
import numpy as np
```

```python
from tqdm import tqdm
```

```python
static_pixel = [0]*255
for (frames,resframe),_ in tqdm(train_loader):
    for i in np.reshape(resframe.numpy(),(-1)):
        static_pixel[int(i*255)] +=1
static_pixel
```
